```python
!pip install tensorflow keras_cv --upgrade --quiet

import keras_cv
import matplotlib.pyplot as plt

model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)

prompts = [
    "A cozy coffee shop in rainy weather, photorealistic, warm lighting, high quality, highly detailed",

    "A woman wearing traditional Habesha dress performing Ethiopian coffee ceremony, indoor setting, natural lighting, realistic, highly detailed",

    "A street in Paris with the Eiffel Tower in the background, people walking, daytime, photorealistic, high detail",

    "A modern city street with cars and pedestrians, sunny day, realistic, high quality",

    "A cat sitting on a wooden table near a window, soft natural light, photorealistic, high detail"
]

for prompt in prompts:
    print("Prompt:", prompt)
    images = model.text_to_image(prompt, batch_size=1)

    plt.figure(figsize=(5,5))
    plt.imshow(images[0])
    plt.axis("off")
    plt.show()
```
